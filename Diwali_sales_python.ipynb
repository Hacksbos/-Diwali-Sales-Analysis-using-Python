```python

# Data
age_sales = df.groupby('Age Group')['Amount'].sum().reset_index().reset_index().sort_values(by='Amount', ascending=False)

# Interactive Bar Chart
fig = px.bar(
    age_sales,
    x='Age Group',
    y='Amount',
    color='Age Group',
    title='🧑‍🤝‍🧑 Age Group wise Total Sales',
text='Amount',
    color_discrete_sequence=px.colors.qualitative.Vivid
)
fig.update_layout(
    xaxis_title='Age Group',
    yaxis_title='Total Sales Amount (₹)',
    title_font_size=18,
    xaxis=dict(showgrid=False),
    yaxis=dict(gridcolor='lightgrey'),
    plot_bgcolor='white',
    legend_title='Age Group'
)
```

```python
# Group & sort data (convert to millions)
state_sales = df.groupby('State')['Amount'].sum().sort_values(ascending=False)/ 1_000_000

# Plot setup
plt.figure(figsize=(10,5))
sns.set_style("whitegrid")
colors = sns.color_palette("viridis", len(state_sales))

# Create bar chart
bars = plt.bar(state_sales.index, state_sales.values, color=colors, edgecolor='black')

# Titles & labels
plt.title("State-wise Total Sales (in Millions)", fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel("State", fontsize=12, color='purple')
plt.ylabel("Total Sales (₹ Million)", fontsize=12, color='purple')
plt.xticks(rotation=45, ha='right')

# Add values (auto-fit inside or above bars)
for bar in bars:
    y = bar.get_height()
    x = bar.get_x() + bar.get_width()/2
    text_y = y/2 if y > state_sales.max()*0.15 else y + state_sales.max()*0.02
    text_color = 'white' if y > state_sales.max()*0.15 else 'black'
    plt.text(x, text_y, f'{y:.2f}M', ha='center', va='center', fontsize=9, color=text_color, fontweight='bold')

# Show chart
plt.tight_layout()
plt.show()
```

```python
# Top 10 Products by Sales (Millions)
top = df.groupby('Product_ID')['Amount'].sum().nlargest(10) / 1_00_00

plt.figure(figsize=(10,5))
sns.set_style("white")  # pure white background (no grid lines)
bars = plt.bar(top.index, top.values, color=sns.color_palette("plasma", 10), edgecolor='none')

# Title & Labels
plt.title("Top 10 Products ID by Sales (in Millions)", fontsize=14, fontweight='bold', color='navy')
plt.xlabel("Product ID"); plt.ylabel("Sales (₹ Million)")
plt.xticks(rotation=45, ha='right')

# Value Labels
for b in bars:
    y=b.get_height(); c='white' if y>top.max()*0.15 else 'black'
    plt.text(b.get_x()+b.get_width()/2, y/2 if y>top.max()*0.15 else y+top.max()*0.02,
             f'{y:.2f}M', ha='center', va='center', fontsize=9, color=c, fontweight='bold')


plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

# Data Prepare
gender_sales = df.groupby('Gender')['Amount'].sum().reset_index()

# Enhanced Interactive Pie Chart
fig = px.pie(
    gender_sales,
    names='Gender',
    values='Amount',
    title='💰 Gender-wise Total Sales Overview',
    hole=0.45,   # Bigger Donut
    color='Gender',
    color_discrete_map={'Female': '#FF6F91', 'Male': '#6FA8DC'}  # Attractive custom colors
)

# Trace Customization
fig.update_traces(
    textposition='inside',
    textinfo='label+percent',
    textfont=dict(size=16, color='white', family='Arial Black'),
    hovertemplate='<b>%{label}</b><br>Sales: ₹%{value:,.0f}<br>Share: %{percent}',
    marker=dict(line=dict(color='white', width=2))
)

# Layout Customization
fig.update_layout(
    title_font=dict(size=24, color='navy', family='Verdana'),
    plot_bgcolor='rgba(0,0,0,0)',
    paper_bgcolor='rgba(245,245,245,1)',
    legend_title='Gender',
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=-0.2,
        xanchor="center",
        x=0.5,
        font=dict(size=14)
    ),
    margin=dict(t=100, b=50, l=50, r=50),
    height=600, width=700  # Bigger Size
)

fig.show()
```
